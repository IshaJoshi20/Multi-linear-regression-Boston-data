```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler #Standardization
from sklearn.model_selection import train_test_split #Splitting dataset
from sklearn.linear_model import LinearRegression # Linear regression model
from sklearn.datasets import load_boston
boston = load_boston()
df = pd.DataFrame(boston.data)
```

```python
#multiple plots in single image
# let's see how data is distributed for every column
plt.figure(figsize=(20,25), facecolor='white')
plotnumber = 1

for column in df:
    if plotnumber<=13 :
        ax = plt.subplot(4,4,plotnumber)
        sns.distplot(df[column]) #plotting 
        plt.xlabel(column,fontsize=20) #label 
        
    plotnumber+=1
plt.tight_layout()
```

```python
from sklearn.datasets import load_boston
X, y = load_boston(return_X_y=True)
```

```python
scaler=StandardScaler()
```

```python
X_scaled=scaler.fit_transform(X)
```

```python
X_scaled
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=111)
```

```python
model=LinearRegression()
```

```python
model.fit(X_train,y_train) #training of the model
```

```python
model.score(X_train,y_train) #training accuracy
```

```python
model.score(X_test,y_test)   #testing accuracy
```

```python
from sklearn.metrics import r2_score
```

```python
predy=model.predict(X_test)
predy
```

```python
y_test
```

```python
r2=r2_score(y_test,predy)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test,predy) #MSE
```

```python
from math import sqrt
```

```python
sqrt(mean_squared_error(y_test,predy)) #RMSE
```
